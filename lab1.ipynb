```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("Black-White:\n")
img_bw = cv2.imread("./KEK.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (15, 15))
plt.imshow(img_bw, cmap = 'gray')
plt.show();

print("\nBGR to RGB:")
img_c = cv2.imread("./KEK.jpg")
img_c_rgb = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (15, 15))
plt.imshow(img_c_rgb)
plt.show();

print("\nBGR to HSV")
img_c_hsv = cv2.cvtColor(img_c, cv2.COLOR_BGR2HSV)
plt.figure(figsize = (15, 15))
plt.imshow(img_c_hsv)
plt.show();

print("\nColor channels for RGB picture:")
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(img_c_rgb.shape, dtype='uint8')
R[:, :, 0] = img_c_rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_c_rgb.shape, dtype='uint8')
G[:, :, 1] = img_c_rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_c_rgb.shape, dtype='uint8')
B[:, :, 2] = img_c_rgb[:, :, 2]
plt.imshow(B)

plt.show()

print("\nColor channels for HSV picture:")
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(img_c_hsv.shape, dtype='uint8')
R[:, :, 0] = img_c_hsv[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_c_hsv.shape, dtype='uint8')
G[:, :, 1] = img_c_hsv[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_c_hsv.shape, dtype='uint8')
B[:, :, 2] = img_c_hsv[:, :, 2]
plt.imshow(B)

plt.show()

print("\nArea of interest for BW picture")
BW_Shp = img_bw.shape
F_Shp = BW_Shp[0]/2
S_Shp = BW_Shp[1]/2

img_crop_bw = img_bw[0 : int(F_Shp), 0 : int(S_Shp)]
plt.imshow(img_crop_bw, cmap = "gray")
plt.show()

cv2.imwrite("./kek.png", img_crop_bw)
```
